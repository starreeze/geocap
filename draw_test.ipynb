```python
import numpy as np  
import matplotlib.pyplot as plt  

# 参数设置
a = 455.7651749644788  # focal_length
c = 0.44088744773694355  # x_offset
m = 3.415401747088463
x_0 = 0.499037241182  # 对称轴 x = x_0
y_0 = 0.45448287718980757
epsilon = 0.01

x_left = np.linspace(0, x_0, 10000)
x = np.linspace(0, 2*x_0, 20000)


for _ in range(5):
    y = ((x_left-c)/(4*a)) ** (1/m) + y_0
    y_symmetric = np.flip(y)

    width = 1 - 2*c
    noise = np.sin((3*np.pi/width)*(x-c) + np.pi) * epsilon
    y = np.concatenate([y, y_symmetric]) + noise
    a = a/2
    c = c-0.05
    epsilon *= 1.5
    # 绘制抛物线
    plt.plot(x, y)
    plt.plot(x, 2*y_0-y)

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 1)  
plt.ylim(0, 1) 

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
a = 0.2
c1 = 0.5
y_0 = 0.5  # 对称轴

x = np.linspace(0, 1, 100)
noise = np.sin(3*np.pi*x) * 0.1


for i in range(5):
    y1 = 4 * a/(1.2**i) * (x-0.5)**2 + c1/(1.2**i) + noise
    y2 = 2 * y_0 - y1  # 对称
    # 绘制抛物线
    plt.plot(x, y1)
    plt.plot(x, y2)


plt.xlim(0, 1)  
plt.ylim(0, 1) 
plt.show()
```
